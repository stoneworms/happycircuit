```python
# 1. 중복된 학교명 + 시도 조합만 골라냄
duplicate_keys = matched.groupby(["학교명_data", "시도_data"]).filter(lambda x: len(x) > 1)

# 2. 그 중에서 시군구 == 행정구_data1 조건만 만족하는 행만 남김
filtered_duplicates = duplicate_keys[duplicate_keys["시군구"] == duplicate_keys["행정구_data1"]]

# 3. 원래 matched에서 중복 아예 아닌 애들만 따로 보존
non_duplicates = matched[~matched.index.isin(duplicate_keys.index)]

# 4. 최종 결과: 비중복 행 + 조건을 만족한 중복 행 합치기
matched = pd.concat([non_duplicates, filtered_duplicates]).sort_index()
```
